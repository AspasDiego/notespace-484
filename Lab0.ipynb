Task 1.1: Write a function called fizzbuzz that takes a number n and returns:\
• "FizzBuzz" if n is divisible by both 3 and 5\
• "Fizz" if n is divisible by 3 only\
• "Buzz" if n is divisible by 5 only\
• The number itself as a string otherwise

```python
def fizzbuzz(n):
    if n % 3 == 0 and n % 5 == 0:
        return "FizzBuzz"
    elif n % 3 == 0:
        return "Fizz"
    elif n % 5 == 0:
        return "Buzz"
    else:
        return str(n)


print(fizzbuzz(15))
```

Task 1.2: Given the following list of dictionaries representing students, write code to:
1. Find the student with the highest grade
2. Calculate the average grade
3. Create a list of names of students who passed (grade >= 60)

```python
students = [
    {"name": "Alice", "grade": 85},
    {"name": "Bob", "grade": 72},
    {"name": "Charlie", "grade": 55},
    {"name": "David", "grade": 91},
    {"name": "Eve", "grade": 68},
]

def highest_grade(students):
    highest_grade = 0

    for student in students:
        if student["grade"] > highest_grade:
            highest_grade = student["grade"]

    return highest_grade

def average_grade(students):
    total_grade = 0
    num_students = len(students)

    for student in students:
        total_grade += student["grade"]
    
    return total_grade/num_students

def passed_students(students):
    pass_list = []

    for student in students:
        if student["grade"] >= 60:
            pass_list.append(student["name"])
    
    return pass_list

print(highest_grade(students))
print(average_grade(students))
print(passed_students(students))
```

# Task 2
Task 2.1: Create a 5x5 matrix with values from 1 to 25. Then:
1. Extract the diagonal elements
2. Calculate the sum of each row
3. Find all elements greater than 12
4. Replace all even numbers with 0

```python
%pip install numpy
```

```python
import numpy as np

#Create a 5x5 matrix with zero values
mat_5x5 = np.zeros((5,5))
num = 1

#Load the matrix with value from 1-25
for i in range(mat_5x5.shape[0]):
    for j in range(mat_5x5.shape[1]):
        mat_5x5[i][j] = num
        num+=1

#Extracting the diagonal elements
diag_elements = []
for i in range(mat_5x5.shape[0]):
    for j in range(mat_5x5.shape[1]):
        if i == j:
            diag_elements.append(int(mat_5x5[i][j]))
print(f"The diagonal elements are :{diag_elements}")

#Calculate sum of each row
sum_row = []
for i in range(mat_5x5.shape[0]):
    sum = 0
    for j in range(mat_5x5.shape[1]):
        sum += mat_5x5[i][j]
    sum_row.append(int(sum))
print(f"The sum of each of the rows of the matrix is: {sum_row}")

#Elements greater than 12
elem_12 = []
for i in range(mat_5x5.shape[0]):
    for j in range(mat_5x5.shape[1]):
        if mat_5x5[i][j] > 12:
            elem_12.append(int(mat_5x5[i][j]))
print(f"The elements greater than 12 are :{elem_12}")

#Replace all even number by 0
for i in range(mat_5x5.shape[0]):
    for j in range(mat_5x5.shape[1]):
        if mat_5x5[i][j] % 2 == 0:
            mat_5x5[i][j] = 0
print(f"The even number replaced by 0 matrix is: \n {mat_5x5}")
```

Task 2.2: Generate 1000 random numbers from a normal distribution with mean=50 and
standard deviation=10. Then calculate and print the actual mean, standard deviation, and the
percentage of values within one standard deviation of the mean.

```python
#Generate 1000 random number
mean = 50
std = 10
nums = np.random.normal(mean,std,1000)

#Mean
mean_val = nums.mean()
print(f"The mean value of the distribution is: {mean_val}")

#Standard Deviation
std_val = nums.std()
print(f"The standard deviation of the distribution is: {std_val}")

#Percetage Values
var_val = nums.var()
print(f"The variance of the distribution is: {var_val}")
```

# Task 3
Task 3.1: Using the students_df DataFrame:
1. Add a column called "total_score" that sums all three subject scores
2. Add a column called "rank" that ranks students by total_score (1 = highest)
3. Create a new DataFrame containing only students who scored above average in all three
subjects


```python
%pip install pandas 
import pandas as pd
```

```python
students_df = pd.read_csv("students.csv")
df = students_df.copy()
df.columns = df.columns.str.strip()
print(df)
```

```python
#Add total_score column
df["total_score"] = (df["math_score"]+df["english_score"]+df["science_score"])

#Rank
df["rank"] = df["total_score"].rank(ascending=False, method="dense").astype(int)
print(f"The dataframe with total scores and rank: \n{df}")

#New datafram of only above average students
avg_score_math = df["math_score"].mean()
avg_score_english = df["english_score"].mean()
avg_score_science = df["science_score"].mean()
df2 = df[
    (df["math_score"] > avg_score_math) &
    (df["english_score"] > avg_score_english) &
    (df["science_score"] > avg_score_science)
]
print(f"\n\nThe dataframe with only above average scoes: \n{df2}")
```

Task 3.2: Calculate and display:
1. The correlation between math_score and science_score
2. The student(s) with the highest improvement potential (biggest difference between their
best and worst subject)
3. The average score for each subject

```python
#Correlation between math_score and science_score
correlation = df["math_score"].corr(df["science_score"])
print(f"Correlation between math_score and science_score: {correlation:.3f}")

#Highest imrpovement potential
df["improvement_potential"] = (
    df[["math_score", "english_score", "science_score"]].max(axis=1) -
    df[["math_score", "english_score", "science_score"]].min(axis=1)
)
max_improvement = df["improvement_potential"].max()
students_with_max_potential = df[
    df["improvement_potential"] == max_improvement
][["student_id", "name", "improvement_potential"]]
print("\nStudent(s) with the highest improvement potential:")
print(students_with_max_potential)

#Average score per subject
average_scores = df[["math_score", "english_score", "science_score"]].mean()
print("\nAverage score for each subject:")
print(average_scores)
```
